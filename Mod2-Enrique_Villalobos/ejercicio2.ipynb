```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats as st
```

```python
#lectura de datos
file_path = os.path.join(os.getcwd(), 'idealista_madrid.csv')
df = np.genfromtxt(
    file_path,
    delimiter=',',
    skip_header=0,
    encoding='utf-8-sig',
    dtype=None,
    names=True
)

```

```python
#APARTADO 3

# Tukey's 
Q1 = np.percentile(df['price'], 25)
Q3 = np.percentile(df['price'], 75)
IQR = Q3 - Q1
lower_bigote = Q1 - 1.5 * IQR
upper_bigote = Q3 + 1.5 * IQR
# Histograma y Boxplot
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Histograma de precios
sns.histplot(df['price'], bins=30, kde=True, ax=axs[0, 0])
axs[0, 0].axvline(lower_bigote, color='green', linestyle='--', label='Tukey lower bigote')
axs[0, 0].axvline(upper_bigote, color='red', linestyle='--', label='Tukey upper bigote')
axs[0, 0].legend()
axs[0, 0].set_title('Estandarizacion Precio con Tukey\'s')
axs[0, 0].set_xlabel('Precio')
axs[0, 0].set_ylabel('Frecuencia')

# Boxplot de precios
sns.boxplot(x=df['price'], ax=axs[0, 1])
axs[0, 1].axvline(lower_bigote, color='green', linestyle='--', label='Tukey lower bigote')
axs[0, 1].axvline(upper_bigote, color='red', linestyle='--', label='Tukey upper bigote')
axs[0, 1].legend()
axs[0, 1].set_title('Boxplot de Precios')
axs[0, 1].set_xlabel('Precio')


z_scores = st.zscore(df['price'])
Q1 = np.percentile(z_scores, 25)
Q3 = np.percentile(z_scores, 75)
IQR = Q3 - Q1
lower_bigote = Q1 - 1.5 * IQR
upper_bigote = Q3 + 1.5 * IQR

# Histograma de Z-Scores
sns.histplot(z_scores, bins=30, kde=True, ax=axs[1, 0])
axs[1, 0].axvline(lower_bigote, color='green', linestyle='--', label='z-score lower bigote')
axs[1, 0].axvline(upper_bigote, color='red', linestyle='--', label='z-score upper bigote')
axs[1, 0].legend()
axs[1, 0].set_title('Estandarizacion Precios con Z-Scores')
axs[1, 0].set_xlabel('Z-Score')
axs[1, 0].set_ylabel('Frecuencia')

# Boxplot de Z-Scores
sns.boxplot(x=z_scores, ax=axs[1, 1])
axs[1, 1].axvline(lower_bigote, color='green', linestyle='--', label='z-score lower bigote')
axs[1, 1].axvline(upper_bigote, color='red', linestyle='--', label='z-score upper bigote')
axs[1, 1].legend()
axs[1, 1].set_title('Boxplot de Z-Scores')
axs[1, 1].set_xlabel('Z-Score')

plt.tight_layout()
plt.show()

```

La transformación de raíz cuadrada mejora significativamente la simetría de las variables (las centra), haciéndolas más adecuadas para análisis que requieren distribuciones más normales. Sin embargo, para casos de asimetría extrema, como los metros cuadrados, pueden considerarse otro tipo de transformaciones. En cuanto a la Kurtosis, la transformación ha sido efectiva para reducir la curtosis en variables con colas largas y valores extremos (precios y los metros cuadrados) lo que ayuda a estabilizar las distribuciones y mejorar los análisis estadísticos. El num de habitaciones y baños tienen una curtosis baja (normal) y la transformación no ha tenido mucho efecto ya que no tiene muchos valores extremos.

```python
# Contraste de hipótesis: Que las casas de X barrio son más baratas de las de Y barrio

# Usamos dos barrios
barrio_X = 'Recoletos'
barrio_Y = 'Castellana'

precios_X = df[df['address'] == barrio_X]['price']
precios_Y = df[df['address'] == barrio_Y]['price']

# Realizar la prueba t de dos muestras
t_stat, p_value = st.ttest_ind(precios_X, precios_Y, alternative='less')

print(f'Estadístico t: {t_stat}')
print(f'Valor p: {p_value}')

# Interpretar el resultado
alpha = 0.05
print('Hipótesis nula: Las casas en {} son más baratas que las de {}'.format(barrio_X, barrio_Y))
if p_value < alpha:
    print(f'Rechazamos la hipótesis nula: Las casas en {barrio_X} son significativamente más baratas que en {barrio_Y}')
else:
    print(f'No podemos rechazar la hipótesis nula: No hay evidencia suficiente para decir que las casas en {barrio_X} son más baratas que en {barrio_Y}')
```

```python
# Filtrar 
precios_3_banos = df[df['baths'] >= 3]['price']
precios_1_2_banos = df[(df['baths'] == 1) | (df['baths'] == 2)]['price']

# Realizar la prueba t de dos muestras
t_stat_banos, p_value_banos = st.ttest_ind(precios_3_banos, precios_1_2_banos, alternative='greater')

print(f'Estadístico t: {t_stat_banos}')
print(f'Valor p: {p_value_banos}')

# Interpretar el resultado
alpha = 0.05
print('Hipótesis nula: Las casas con 3 o más baños son más caras que las casas con 1-2 baños')
if p_value_banos < alpha:
    print(f'Rechazamos la hipótesis nula: Las casas con 3 o más baños son significativamente más caras que las casas con 1-2 baños')
else:
    print(f'No podemos rechazar la hipótesis nula: No hay evidencia suficiente para decir que las casas con 3 o más baños son más caras que las casas con 1-2 baños')
```
